```python
from sqlalchemy import create_engine
import pandas as pd
from config import db_password
```

```python
# Files to load
UserInfo_data_to_load = "Resources/CarDetailsV3_UserInfo.csv"
EngineInfo_data_to_load = "Resources/CarDetailsV3_CarEngineInfo.csv"
CarYear_data_to_load = "Resources/CarDetailsV3_CarYear.csv"
```

```python
# Read the data files and store it in a pandas DataFrame.
UserInfo_df = pd.read_csv(UserInfo_data_to_load)
UserInfo_df.head(10)
```

```python
EngineInfo_df = pd.read_csv(EngineInfo_data_to_load)
EngineInfo_df.head(10)
```

```python
CarYear_df = pd.read_csv(CarYear_data_to_load)
CarYear_df.head(10)
```

```python
# Create the connection string using the password from the config.py file
db_string = f"postgres://postgres:{db_password}@127.0.0.1:5432/Car_Dekho_Prediction"
```

```python
# Create the database engine
engine = create_engine(db_string)
```

```python
# Import the CarYear file to PostGreSQL
CarYear_df.to_sql(name='CarYear', con=engine)
```

```python
# Import the EngineInfo file to PostGreSQL
EngineInfo_df.to_sql(name='CarEngineInfo', con=engine)
```

```python
# Import the UserInfo file to PostGreSQL
UserInfo_df.to_sql(name='UserInfo', con=engine)
```
